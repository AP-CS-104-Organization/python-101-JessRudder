## Chapter:9 Plotting and Visualization

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data1=np.arange(10)
data2=np.array([0,2,2,6,4,10,6,14,8,18])
data3=np.array(np.random.randint(0,18,20))
data4=np.array(np.random.randn(1000))
```

```python
# setting  font style:
from matplotlib.font_manager import FontProperties
font=FontProperties()
font.set_family('serif')# 'serif', 'sans-serif', 'cursive', 'fantasy', or 'monospace'
font.set_name('Times New Roman')
font.set_style('italic')#  'normal', 'italic' or 'oblique'.
font.set_size(40)
font.set_weight('light')#'light', 'normal', 'regular', 'book', 'medium', 'roman', 'semibold', 'demibold', 'demi', 'bold', 'heavy', 'extra bold', 'black'
```

```python
## General Plot with legends, title, labels,linestyle,color, marker 
#fig=plt.figure()
plt.plot(data1,'k-',label='default')
#plt.plot(data1,linestyle='-',color='k')
plt.plot(data2,'r*--',label='steps')
#plt.plot(data2,linestyle='--',color='r',marker='o')
plt.legend(loc='best',fontsize=12)
plt.title('My first matplotlib plot',fontsize=12,style='italic',fontweight='bold')
plt.xlabel('Stages',fontsize=12)
plt.ylabel('Value',fontproperties=font)
#Changing the size of figure, axes, grid,legend,xtick,ytick
plt.rc('figure',figsize=(10,10))
#plt.rc('axes',axessize=12)
```

```python
## Subplots 

fig=plt.figure()
fig.subplots_adjust(hspace=0.4, wspace=0.4)
ax1=fig.add_subplot(2,2,1)
ax1.plot(data1,'k-',label='Default')
plt.legend()
plt.xlabel('Stages')
plt.ylabel('Values')
ax2=fig.add_subplot(2,2,2)
ax2.plot(data2,'r--*',label='Step')
plt.legend()
plt.xlabel('Stages')
plt.ylabel('Values')
##Scatter plot
ax3=fig.add_subplot(2,2,3)
ax3.scatter(np.arange(20),data3,label='Scater',color='g',s=25,marker='*',alpha=0.9)
plt.legend(loc='best')
plt.xlabel('Stages')
plt.ylabel('Values')
## Histogram plot
ax4=fig.add_subplot(2,2,4)
ax4.hist(data4,bins=10,alpha=0.9,histtype='stepfilled',color='g',edgecolor='r',label='Aalu')
plt.legend()#loc='best')
plt.xlabel('Stages')
plt.ylabel('Values')
```

###### Ticks: xticks, yticks, xlim, sticklabels

```python
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
ax.plot(data1,'k-',label='default')
#plt.plot(data1,linestyle='-',color='k')
ax.plot(data2,'r*--',label='steps')
#plt.plot(data2,linestyle='--',color='r',marker='o')
ax.legend(loc='best',fontsize=12)
ax.set_title('My first matplotlib plot',fontsize=12,style='italic',fontweight='bold')
ax.set_xlabel('Stages',fontsize=12)
ax.set_ylabel('Value',fontproperties=font)
#Changing the size of figure, axes, grid,legend,xtick,ytick
#plt.rc('figure',figsize=(10,10))
#plt.rc('axes',axessize=12)
ticks=ax.set_xticks([0,2,4,6,8])
labels=ax.set_xticklabels(['zero','two','four','six','eight'],rotation=30,fontsize=20)
```

## Plotting with Seaborn

```python
# Importing tips csv file
tips=pd.read_csv(r'E:\Machine Learning\python\Python for data analysis wes\examples\tips.csv')
```

```python
tips.head(5)
```

```python
tips['tip_pct']=tips['tip']/(tips['total_bill']-tips['tip'])
```

```python
tips.head()
```

##### Bar plot

```python
import seaborn as sns
sns.set(style="whitegrid")
sns.barplot(x='tip_pct',y='day',hue='time',data=tips,orient='h')
```

```python
sns.barplot(x='tip_pct',y='day',hue='time',data=tips)
```

```python
sns.barplot(x='time',y='tip',hue='day',data=tips,order=['Dinner','Lunch'])
```

```python
from numpy import median
sns.barplot(x='day',y='tip',data=tips,estimator=median)
```

```python
# Showing standard deviation of observation instead fo confidence interval
sns.barplot(x='day',y='tip',data=tips,ci='sd',capsize=0.4)
```

##### Facet grids and categorical data

```python
sns.catplot(x='day',y='tip_pct',hue='time',col='smoker',kind='bar',data=tips)
```

```python
sns.catplot(x='day',y='tip_pct',row='time',col='smoker',kind='bar',data=tips[tips.tip_pct<1])
```

```python
sns.catplot(x='tip_pct',y='day',kind='box',data=tips[tips.tip_pct<0.5])
# show median,quartiles and outliers :can be a efffective visualization type
```

##### Histograms and densiy plots

```python
tips['tip_pct'].plot.hist(bins=100)
```

```python
tips['tip_pct'].plot.density()
```
